### Customer Behaviour Trends

```python
#Repeat customers
repeat_customer = trends_df.groupby('Customer ID')[['Frequency of Purchases', 'Previous Purchases']].agg(['sum','max'])
repeat_customer
```

Most active customers likely fall into Weekly or Fortnightly purchase frequency, indicating a habit of frequent engagement with the business.
Less frequent purchasers (e.g., Annually) might represent lower engagement or seasonal customers

```python
#Demographics 
demo = trends_df.groupby(['Age','Gender','Location'])['Purchase Amount (USD)'].agg(['sum','max','mean'])
demo
```

Insights and Trends:  
Young Customers Are Key in Specific Locations:

Younger customers (e.g., Age 18) show significant activity in Kansas but low engagement in Kentucky or Maryland.
Target marketing campaigns at young customers in high-potential regions like Kansas.
Older Customers Show High Value but Region-Specific Trends:

Older customers (Age 70) contribute high activity in Texas and Vermont, with Vermont showing consistent high-value interactions (mean = 69.67).
Consider loyalty or premium service offerings tailored for older customers in these states.
Gender Differences by Location:

Female customers tend to be highly active in Kansas and Illinois, indicating regions where campaigns targeted at women could be impactful.
Male customers dominate high-value interactions in Vermont and Texas, which could guide gender-specific strategies.
Regional Disparities Exist:

High-potential regions like Kansas (for young customers) and Vermont (for older customers) contrast with lower-engagement states like South Dakota and Kentucky.
Focus resources on high-activity states for better ROI and investigate barriers to engagement in low-activity regions.

```python
# subscription Impact
subs_status=trends_df.groupby('Subscription Status')[['Purchase Amount (USD)','Frequency of Purchases']].agg(['sum','max'])
subs_status
```

Non-Subscribers Spend More:Non-subscribers contribute 73% of total purchase value ($170,436) compared to subscribers ($62,645).

Weekly Purchases Dominate:Both groups show Weekly purchases as the most frequent pattern.

Diverse Frequencies for Non-Subscribers:Non-subscribers shop across a wider range of frequencies (e.g., Bi-Weekly, Quarterly) compared to subscribers.

Subscriber Value Concerns:Subscribers spend less overall, suggesting the subscription model may not be maximizing customer lifetime value.

Equal Max Transaction ($100):Both groups have the same highest single transaction value, indicating no difference in spending power per purchase.


Action Points:  
Enhance subscription benefits to boost value and spending.
Use weekly purchase trends to optimize campaigns.
Target non-subscribers for conversion and engagement strategies.

### Shipping and Logistics

```python
#shipping type
ship_type = trends_df.groupby('Shipping Type')['Purchase Amount (USD)'].agg(['sum','mean','max'])
ship_type
```

Most Popular Shipping Types (by Total Purchase Amount):Free Shipping leads with the highest total purchase value ($40,777), followed by Express ($39,067) and Store Pickup ($38,931).

Highest Average Purchase Amount:2-Day Shipping has the highest average purchase value ($60.73), slightly ahead of Express ($60.47).

Equal Max Purchase Across All Types:The maximum single transaction value is consistent ($100) for all shipping types.

Lower Averages for Some Types:Standard Shipping and Next Day Air have the lowest average purchase values ($58.46 and $58.63, respectively).

Actionable Takeaways  
Promote Free Shipping: It leads in total purchase value, indicating strong customer preference.
Optimize Pricing for 2-Day Shipping and Express: These offer the highest average value, suggesting premium service appeal.
Reassess Standard and Next Day Air: Their lower averages may indicate they lack perceived value compared to other options.


### Product Reviews

```python
#customer satisfaction 
cust_sastsfn = trends_df.groupby('Review Rating')[['Item Purchased','Category']].agg(['sum','max'])
cust_sastsfn
```

Outerwear Leads: Outerwear dominates maximum purchases across all review ratings.  
T-shirt Popularity: T-shirts are the most frequently purchased item in maximum value.  
Diverse Preferences: Sum values show variety with Clothing, Accessories, Footwear, and Outerwear.  
High Ratings = Outerwear: Customers with higher ratings prefer outerwear purchases.  
Accessory Add-ons: Accessories are common in mid-range review ratings (3.0 to 4.0).  
Opportunities in Bundling: Combine T-shirts and Outerwear for promotional sales

Promote Outerwear: Focus on campaigns highlighting outerwear for high-value and high-rating customers.  
Bundle T-shirts: Create bundles featuring T-shirts to leverage their universal popularity.  
Target Mid-Rating Buyers: Offer discounts or personalized deals on Accessories to mid-range rating customers (3.0 to 4.0)  
Upsell Outerwear: Use positive reviews to upsell premium outerwear to high-rating customers.  
Seasonal Promotions: Align marketing efforts for Outerwear with seasonal demands to maximize engagement.

```python
df = trends_df.sample(10)
```

```python
data = df['Previous Purchases']
plt.hist(data,bins=5,color='orange',edgecolor='black');
plt.title('Previuos Purches')
```

```python
total_sale_by_product = df.groupby('Category')['Purchase Amount (USD)'].sum()
total_sale_by_product.plot(kind = 'bar', color='teal')
plt.title('Total sale by products')
```

```python
#Sales over Rating
sales_rating = df.groupby('Review Rating')['Purchase Amount (USD)'].sum().reset_index()
plt.plot(sales_rating['Review Rating'],sales_rating['Purchase Amount (USD)']);
plt.title('Sales over Rating')
```
